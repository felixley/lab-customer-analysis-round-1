```python
import pandas as pd
import numpy as np
```

```python
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')
```

```python
file1_columns = [item.upper() for item in file1.columns]
file2_columns = [item.lower() for item in file2.columns]
file3_columns = [item.lower() for item in file2.columns]
```

```python
file2 = file2[['Customer', 'ST', 'GENDER', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type',  'Vehicle Class','Total Claim Amount']]
```

```python
file3 = file3[['Customer', 'State','Gender',  'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class','Total Claim Amount', ]]
```

```python
column_names = ['CUSTOMER','STATE','GENDER','EDUCATION','CUSTOMER LIFETIME VALUE','INCOME','MONTHLY PREMIUM AUTO','NUMBER OF OPEN COMPLAINTS','POLICY TYPE','VEHICLE CLASS','TOTAL CLAIM AMOUNT']
data = pd.DataFrame(columns = column_names)
data = pd.concat([file1, file2, file3], axis = 0)
data.shape
```

```python
data.select_dtypes(np.object)
```

```python
data.drop(['Education', 'Number of Open Complaints'],  axis =1, inplace = True)
```

```python
data['Customer Lifetime Value'].replace({'%':''},regex = True)
```

```python
data['Customer Lifetime Value'] = pd.to_numeric(data['Customer Lifetime Value'], errors = 'coerce')
```

```python
data['Customer Lifetime Value'] = [x*100 for x in data['Customer Lifetime Value']]
```
